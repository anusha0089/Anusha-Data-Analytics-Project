Cleaning the data

```python
# Convert the 'Date' column to datetime format
df['SleepDay'] = pd.to_datetime(df['SleepDay'], format='%m/%d/%Y %I:%M:%S %p')

# Convert the 'Date' column to date only
df['SleepDay'] = df['SleepDay'].dt.strftime('%Y-%m-%d')

# Print the Data
```
