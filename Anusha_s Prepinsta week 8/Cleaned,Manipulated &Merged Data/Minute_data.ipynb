# ***MERGING DATA SETS***

```python
import pandas as pd

# Display all columns without truncation
pd.set_option('display.max_columns', None)

# Load related dataset from URL into multiple DataFrame
url_1 = '/content/minute_METs_Narrow_cleaned.csv'
met = pd.read_csv(url_1, encoding='unicode_escape')
url_2 = '/content/minute_calories_narrow_cleaned.csv'
calories = pd.read_csv(url_2, encoding='unicode_escape')
url_3 = '/content/minute_intensity_narrow_cleaned.csv'
intensity = pd.read_csv(url_3, encoding='unicode_escape')
url_4 = '/content/minute_Sleep_cleaned.csv'
sleep = pd.read_csv(url_4, encoding='unicode_escape')
url_5 = '/content/minute_Steps_Narrow_cleaned.csv'
steps = pd.read_csv(url_5, encoding='unicode_escape')

```

```python
print(met.sample(1))
print(calories.sample(1))
print(intensity.sample(1))
print(sleep.sample(1))
print(steps.sample(1))
```

```python
calories.head()
```

```python
intensity.head()
```

```python
met.head()
```

```python
sleep.head()
```

```python
# Drop the column named 'column_name' from the DataFrame
sleep.drop(columns=['logId'], inplace=True)
```

```python
# Rename columns in the 'sleep' DataFrame to match the column names in 'met_calories_intensity'
sleep.rename(columns={'value': 'Sleep'}, inplace=True)
```

```python
sleep.head()
```

```python
steps.head()
```

```python
# Merge DataFrames on specified columns
met_calories = pd.merge(met, calories, on=['Id', 'ActivityDay', 'ActivityMinute'], how='left')

# Display a random sample of 5 rows from the merged DataFrame
met_calories.head(5)
```

```python
# Merge DataFrames on specified columns
met_calories_intensity = pd.merge(met_calories, intensity, on=['Id', 'ActivityDay', 'ActivityMinute'], how='left')

# Display a random sample of 5 rows from the merged DataFrame
met_calories_intensity.head(5)
```

```python
# Merge DataFrames on specified columns
final = pd.merge(met_calories_intensity, steps, on=['Id', 'ActivityDay', 'ActivityMinute'], how='left')

# Display a head sample of 5 rows from the merged DataFrame
final.head(5)
```

```python
# Create a list of column names to drop
columns_to_drop = [col for col in final.columns if col.startswith('Unnamed:')]

# Drop the columns from the DataFrame
final.drop(columns=columns_to_drop, inplace=True)
```

```python
final.head()
```

```python
final.to_csv('minute_data.csv', index=False)
```
