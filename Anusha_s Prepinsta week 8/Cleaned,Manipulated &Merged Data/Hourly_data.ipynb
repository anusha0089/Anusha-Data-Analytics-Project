IMPORTING LIBRARIRS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

UNDERSTANDING DATA

```python
pd.set_option('display.max_columns', None)
url = '/content/hourlyCalories_merged.csv'
df = pd.read_csv(url, encoding='unicode_escape')
df.head()
```

SEPERATING DATE AND TIME

```python
# Convert the 'ActivityHour' column to datetime format and extract date and time components
df['ActivityHour'] = pd.to_datetime(df['ActivityHour'], format='%m/%d/%Y %I:%M:%S %p')
df['ActivityDay'] = df['ActivityHour'].dt.strftime('%Y-%m-%d')  # Extract date in YYYY-MM-DD format
df['ActivityHour'] = df['ActivityHour'].dt.strftime('%H:%M')    # Extract time in HH:MM format

# Rearrange columns to have 'ActivityDay' next to 'ActivityHour'
df = df[['ActivityHour', 'ActivityDay'] + [col for col in df.columns if col not in ['ActivityHour', 'ActivityDay']]]

# Print the DataFrame to see the changes
print(df.head())
```

CHECKING MISSING VALUES

```python
# Check for missing values in the DataFrame
missing_values = df.isnull()

# Count the number of missing values in each column
missing_counts = missing_values.sum()

# Print the missing value counts
print("Missing value counts:")
print(missing_counts)
```

CLEAN DATA

```python
df.to_csv('cleaned_hourly_calories.csv', index=False)
```

MERGING DATA

```python
# Display all columns without truncation
pd.set_option('display.max_columns', None)

# Load related dataset from URL into multiple DataFrame
url_1 = '/content/cleaned_hourly_calories.csv'
calories = pd.read_csv(url_1, encoding='unicode_escape')
url_2 = '/content/cleaned_hourly_intensity.csv'
intensity = pd.read_csv(url_2, encoding='unicode_escape')
url_3 = '/content/cleaned_hourly_steps.csv'
steps = pd.read_csv(url_3, encoding='unicode_escape')
```

```python
# Display random row from each dataframe
print(calories.sample())
print(intensity.sample())
print(steps.sample())
```

```python
# Merge DataFrames on specified columns
calories_intensity = pd.merge(calories, intensity, on=['Id', 'ActivityDay', 'ActivityHour'], how='left')

# Display a head sample of 5 rows from the merged DataFrame
calories_intensity.head(5)
```

```python
# Merge DataFrames on specified columns
final = pd.merge(calories_intensity, steps, on=['Id', 'ActivityDay', 'ActivityHour'], how='left')

# Display a head sample of 5 rows from the merged DataFrame
final.head(5)
```

```python
final.duplicated().sum()
```

```python
final.isnull().sum()
```

```python
final.to_csv('hourly_data.csv', index=False)
```
